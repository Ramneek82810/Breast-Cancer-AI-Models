```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
```

```python
cancer= load_breast_cancer()
df= pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
df['target']= cancer.target
df['target_names']= df['target'].apply(lambda x: cancer.target_names[x])
```

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
feature_dropdown = widgets.Dropdown(
    options=df.columns[:-2],
    description='Features:',
)

target_checkbox = widgets.Checkbox(
    value=True,
    description='Show Target',
)

def update_plot(feature, show_target):
    plt.figure(figsize=(10,6))
    sns.histplot(df[feature], kde=True,color='blue')
    if show_target:
        plt.title(f'Distribution of {feature} by Target')
        sns.histplot(df[df['target']==0][feature], kde=True, color='red', label='Malignant')
        sns.histplot(df[df['target']==1][feature], kde=True, color='green', label='Benign')
        plt.legend()
    else:
        plt.title(f'Distribution of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Frequency')
        plt.show()
        
```

```python
widgets.interactive(update_plot, feature=feature_dropdown, show_target= target_checkbox)
```
